```python
import pandas as pd
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fib_rec = pd.read_csv('fib_recursive.txt')
fib_rec_mod = pd.read_csv('fib_recursive_mod.txt')
fib_iterative_mod = pd.read_csv('fib_iterative_mod.txt')
fib_iterative = pd.read_csv('fib_iterative.txt')
fib_matrix = pd.read_csv('fib_matrix.txt')
fib_matrix_mod = pd.read_csv('fib_matrix_mod.txt')
```

```python
def plot_graph(model, title):
    plt.plot(model['Fib_number'], model['Time_spent'])
    plt.title(title)
    plt.xlabel('Fibonacci Number')
    plt.ylabel('Time in Seconds')
    plt.show()
```

```python
plot_graph(fib_rec, 'Fibonacci Recursive')
```

```python
plot_graph(fib_rec_mod, 'Fibonacci Recursive (Mod)')
```

```python
plot_graph(fib_iterative, 'Fibonacci Iterative')
```

```python
plot_graph(fib_iterative_mod, 'Fibonacci Iterative (Mod)')
```

```python
plot_graph(fib_matrix, 'Fibonacci Matrix')
```

```python
plot_graph(fib_matrix_mod, 'Fibonacci Matrix (Mod)')
```
